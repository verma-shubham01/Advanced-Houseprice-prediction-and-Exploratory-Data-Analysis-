# Feature  Engineering


## 1. Missing Values 
## 2. Temporal variable 
## 3. Categorical Variable remove rare labels
## 4.  Standardization of the values of variables to the same range 

```python
dataset1=pd.read_csv("data.csv")
dataset1.head()
```

## Missing Value

```python
# make list of missing features
# categorical feature that are missing
features_nan=[features for features in dataset1.columns if dataset1[features].isnull().sum()>1 and dataset1[features].dtype=='O']
#print features name and percentage of missing value
for feature in features_nan:
    print(feature, np.round(dataset1[feature].isnull().mean(),4), '% missing  values')
```

```python
# replace missing value with new label
def replace_cat_feature(dataset1,features_nan):
    data=dataset1.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data
dataset1=replace_cat_feature(dataset1,features_nan)

dataset1[features_nan].isnull().sum()
```

```python
dataset1.head()
```

```python
#Numerical variables that contains missing values 
numerical_with_nan=[features for features in dataset1.columns if dataset1[features].isnull().sum()>1 and dataset1[features].dtype !='O']
#print features name and percentage of missing value
for feature in numerical_with_nan:
    print(feature, np.round(dataset1[feature].isnull().mean(),4), '% missing  values')
```

```python
# Replacing the numerical missing value 
for feature in numerical_with_nan:
    # due to outliers ,replace by using median
    median_value=dataset1[feature].median()
    
    #create a new feature to capture new value 
    dataset1[feature+'nan']=np.where(dataset1[feature].isnull(),1,0)
    dataset1[feature].fillna(median_value,inplace=True)
    
dataset1[numerical_with_nan].isnull().sum()
    
```

```python
dataset1.head(50)
```

```python
#Temporal variables(date time variable)
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset1[feature]=dataset1['YrSold']-dataset1[feature]
```

```python
dataset1.head()
```

```python
dataset1[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

# Numerical Variables
since numerical variables are skewed perform log normal distribution

```python
dataset1.head()
```

```python
num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in num_features:
    dataset1[feature]=np.log(dataset1[feature])
dataset1.head()
```

# Handling rare  categorical Feature
remove categorical variable that are present less than 1% of the observations

```python
categorical_features=[feature for feature in dataset1.columns if dataset1[feature].dtype=='O']
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset1.groupby(feature)['SalePrice'].count()/len(dataset1)
    temp_df=temp[temp>.01].index
    dataset1[feature]=np.where(dataset1[feature].isin(temp_df),dataset1[feature],'Rare_var')
dataset1.head()
```

# Change Categorical features to Numerical features

```python
for feature in categorical_features:
    labels_ordered=dataset1.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset1[feature]=dataset1[feature].map(labels_ordered)
dataset1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
sc=LabelEncoder()
for feature in categorical_features:
    dataset1[feature]=sc.fit_transform(dataset1[feature])
dataset1.head()
```

# Feature scaling

```python
feature_scale=[feature for feature in dataset1.columns if feature not in ['Id', 'SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(dataset1[feature_scale])
```

```python
scaler.transform(dataset1[feature_scale])
```

```python
#transform the train and test set  and add on the Id ,SalePrice Variables
dataset2=pd.concat([dataset1[['Id','SalePrice']].reset_index(drop=True),
                  pd.DataFrame(scaler.transform(dataset1[feature_scale]), columns=feature_scale)],axis=1)
```
